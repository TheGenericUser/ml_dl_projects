```python
def changeToStr(list_string):
    L = []
    string = ''
    for x in list_string:
        if x == '[' or x == "'" or x == ",":
            continue
        if x == ' ':
            L.append(string)
            string = ''
            continue
        if x == ']':
            return L
        string += x
```
