```python
ratings = pd.read_csv('book_rec/Ratings.csv',encoding="latin-1")
```

```python
ratings.rename(columns={'User-ID':'id','ISBN':'isbn','Book-Rating':'rating'}, inplace=True)
```

```python
mask = ratings['id'].value_counts()>200
```

```python
mask_index = mask[mask].index
```

```python
mask[mask]
```

```python
mask_index
```

```python
ratings = ratings[ratings['id'].isin(mask_index)]
```

```python
ratings.shape
```

```python
ratings
```

```python
rating = pd.merge(ratings, books, on='isbn', how='inner')
```

```python
ratings.groupby('title')
```

```python
rating
```

```python
rating_per_book = rating.groupby('title')['rating'].count().reset_index()
rating_per_book.rename(columns={'rating':'ratings_per_book'}, inplace=True)
```

```python
final_rating = pd.merge(rating, rating_per_book, on='title')
```

```python
final_rating = final_rating[final_rating['ratings_per_book'] >= 30]
```

```python
final_rating
```

```python
final_rating.drop_duplicates(['id','title'], inplace=True)
```

```python
final_pivot = final_rating.pivot_table(columns='id', index='title', values='rating')
```

```python
final_pivot.fillna(0.0, inplace=True)
```

```python
from scipy.sparse import csr_matrix
```

```python
book_sparse = csr_matrix(final_pivot)
```

```python
from sklearn.neighbors import NearestNeighbors
model = NearestNeighbors(algorithm = 'brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions = model.kneighbors(final_pivot.iloc[556,:].values.reshape(1,-1), n_neighbors=6)
```
