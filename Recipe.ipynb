```python
def clean_brackets(text):
    keys = list(map_dict.keys())
    text = [t for t in text.split(",")]
    for i in range(len(text)):
        for key in keys:
            if key in text[i]:
                text[i] = map_dict.get(key)
    return ",".join(text)
clean_brackets('karela (bitter gourd/ pavakkai) - deseeded,onion - thinly sliced,tablespoon gram flour (besan),tablespoon amchur (dry mango powder)')
```
