```python
import ast
```

```python
def convert(json):
    tag_list = []
    json = ast.literal_eval(json)
    for json_dict in json:
        tag_list.append(json_dict['name'])
    return tag_list
```

```python
def convert2(json):
    tag_list = []
    json = ast.literal_eval(json)
    for json_dict in json:
        tag_list.append(json_dict['iso_3166_1'])
    return tag_list
```

```python
df['production_countries'] = df['production_countries'].apply(convert2)
```

```python
def convert_cast(json):
    tags_list = []
    json = ast.literal_eval(json)

    for i, json_dict in enumerate(json):
        if i >= 6:
            break
        tags_list.append(json_dict['name'])
    return tags_list
```

```python
df['cast'] = df['cast'].apply(convert_cast)
```

```python
df['crew'].iloc[0]
```

```python
def convert_crew(json):
    tags_list = []
    json = ast.literal_eval(json)

    for json_dict in json:
        if json_dict['job'] ==  'Director':        
            tags_list.append(json_dict['name'])
        if json_dict['job'] ==  'Producer':
            tags_list.append(json_dict['name'])
    return tags_list
```

```python
df['crew'] = df['crew'].apply(convert_crew)
```
